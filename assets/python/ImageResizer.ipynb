```python
from PIL import Image
```

```python
img = Image.open('../img/posts/2019/11/220px-Nightfall_cover.jpg')
width, height = img.size

if width < height:
    left = 0 
    right = width
    top = (height - width) / 2
    bottom = height - top
elif width > height:
    top = 0
    bottom = height
    left = (width - height) / 2
    right = width - left

img.crop((left, top, right, bottom))
```

```python
img.save()
```
